```python
import numpy as np
import pandas as pd
import plotly.express as px
```

```python
etl = pd.read_csv("/content/sample_data/etl.csv")
```

## Prompt asks to "visually distinguish each marketâ€™s respective trends over time". However, within markets there are multiple neighborhoods with the same time stamp. Therefore, in order to get a clearer market trend I used distinct delivery window ID's for x-axis tick marks.

```python
fig = px.line(etl, x="delivery_window_id", y="no_orders_placed", color='market_name')
fig.show()
```

Highest demand by huge margin comes from NY market. SF and Phili demand are most comparable to each other. 

```python
fig = px.line(etl, x="delivery_window_id", y="no_unique_avail_shoppers", color='market_name')
fig.show()
```

There is not a 1:1 coverage for demand since the max amount of supply for NY delivery windows is 62 available shoppers. This means there has to be shoppers picking up close to 5 orders to hit demand-supply equilibrium in that market. The gap between demand and supply is not as huge in SF/Phili markets.

```python
fig = px.line(etl, x="delivery_window_id", y="no_unique_fulfilled_shoppers", color='market_name')
fig.show()
```

Most reduction in supply happened in NY market.
