#### Transforming y-value (classification value) into binary set of values.
#### For values 0-->0, for values >1 -->1 representing no cancellation, and yes cancellation

```python
usfares2014['X_Cancel_Binary'] = np.sign(usfares2014['X_Cancel'])
```

```python
usfares2014['XL_Cancel_Binary'] = np.sign(usfares2014['XL_Cancel'])
```

```python
usfares2014.head
```

```python
#Deleting X_Cancel
usfares2014_=usfares2014.drop('X_Cancel', axis=1)
usfares2014_
```

```python
#Deleting XL_Cancel
usfares2014__=usfares2014_.drop('XL_Cancel', axis=1)
usfares2014__.head()
```

#### Specifying Training and Test data further to feed into M.L. Logisitc Regressor. 

#### Training models with 80% of data points.
#### Testing models with 20% data points.

#### Training and Testing data further split up into x and y components.

#### x is all data without cancel column (X/XL)
#### y is all data with cancel column (X/XL)

---
#### One model will be made for X_Uber and another one will be made for XL_Uber. 
##### Also, for feature selection what contributes most to target variable is not Safety, but set fares.
##### So Safety will be dropped from final dataset along with Min Fare. This feature (Min Fare) is a calculation based on other fares we are using as predictors.
##### Thus, it does not contribute to our target value as well.
##### Lastly, city names need to be taken out of the dataset since these are non-numerical. Fare contributors will be traced back to cities via indices.

```python
X_usfares2014 = usfares2014__[['X_Base Fare', 'X_Minute', 'X_Mile', 'X_Cancel_Binary']]
X_usfares2014.head()
```

```python
XL_usfares2014 = usfares2014__[['XL_Base Fare', 'XL_Minute', 'XL_Mile', 'XL_Cancel_Binary']]
XL_usfares2014.head()
```

Data is now ready to be split into Training and Test datasets for Training data to create Logistic Regression model and performance checked with Test set data.

```python
X_x = X_usfares2014.drop('X_Cancel_Binary', axis = 1)
X_y = X_usfares2014['X_Cancel_Binary']

XL_x = XL_usfares2014.drop('XL_Cancel_Binary', axis = 1)
XL_y = XL_usfares2014['XL_Cancel_Binary']
```

X and XL data will not be split into Training and Test sets.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression

X_train_X, X_test_X, y_train_X, y_test_X = train_test_split(X_x, X_y, test_size=0.2)
X_train_XL, X_test_XL, y_train_XL, y_test_XL = train_test_split(XL_x, XL_y, test_size=0.2)
```

```python
#Check to see if data was split correctly.
print X_train_X.shape, y_train_X.shape
print X_test_X.shape, y_test_X.shape

print X_train_XL.shape, y_train_XL.shape
print X_test_XL.shape, y_test_XL.shape
```

```python
#Now that data is split correctly, can feed data into Logistic Regression object.

logreg_X = LogisticRegression()
logreg_X.fit(X_train_X, y_train_X)
```

```python
logreg_XL = LogisticRegression()
logreg_XL.fit(X_train_XL, y_train_XL)
```

```python
y_pred_X = logreg_X.predict(X_test_X)
print('Accuracy of logistic regression classifier on Uber_X test set: {:.2f}'.format(logreg_X.score(X_test_X, y_test_X)))
```

```python
y_pred_XL = logreg_XL.predict(X_test_XL)
print('Accuracy of logistic regression classifier on Uber_XL test set: {:.2f}'.format(logreg_XL.score(X_test_XL, y_test_XL)))
```

Accuracy for Uber_XL higher compared to Uber_X due to balanced data set (similar amount of non-cancellations as there were cancellations). 
Given, there was not much variability/chance for there to be misclassification since data had an even amount of each classification value to train the model well. There was not much swing between cancellations and no cancellations.

More Model Performance Metrics:

```python
from sklearn.metrics import confusion_matrix
confusion_matrix_X = confusion_matrix(y_test_X, y_pred_X)
print(confusion_matrix_X)
```

::The result is telling us for Uber_X model, we have 23 correct predictions and 1 incorrect prediction. Not bad.

```python
from sklearn.metrics import confusion_matrix
confusion_matrix_XL = confusion_matrix(y_test_XL, y_pred_XL)
print(confusion_matrix_XL)
```

::The result is telling us for Uber_XL model, we have 24 correct predictions and 0 incorrect prediction. Not bad at all.

Computing Precision, Recall, F-Score and Support

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_X, y_pred_X))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test_XL, y_pred_XL))
```

#### Borrowing from scikit learn:

The precision is the ratio tp / (tp + fp) where tp is the number of true positives and fp the number of false positives. The precision is intuitively the ability of the classifier to not label a sample as positive if it is negative.

The recall is the ratio tp / (tp + fn) where tp is the number of true positives and fn the number of false negatives. The recall is intuitively the ability of the classifier to find all the positive samples.

The F-beta score can be interpreted as a weighted harmonic mean of the precision and recall, where an F-beta score reaches its best value at 1 and worst score at 0.

The F-beta score weights the recall more than the precision by a factor of beta. beta = 1.0 means recall and precision are equally important.

The support is the number of occurrences of each class in y_test.

::Some comments about the Classification Reports for Uber_X and Uber_XL data:

There is data imbalance for Uber_X since most of the fares resulted in non-cancellations 23/24 = 95.8%.
For Uber_XL, the data was more balanced - almost 50% for cancellations and 50% for pick-up's. Therefore Uber_XL model performed better than Uber_X since the data quality was better.

Now, and finally, the ROC Curve:

```python
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
import matplotlib.pyplot as plt 

logit_roc_auc = roc_auc_score(y_test_X, logreg_X.predict(X_test_X))
fpr, tpr, thresholds = roc_curve(y_test_X, logreg_X.predict_proba(X_test_X)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for Uber_X')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

```python
logit_roc_auc = roc_auc_score(y_test_XL, logreg_XL.predict(X_test_XL))
fpr, tpr, thresholds = roc_curve(y_test_XL, logreg_XL.predict_proba(X_test_XL)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic for Uber_XL')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

::Some comments on ROC Curve for Uber_X and Uber_XL:
            
The ROC Curve is another common tool used to assess how well a binary classifier did. 
The dotted red line represents a purely random classifier.
The ideal situation is for the Logisitc Regression curve to be in the top left corner.

As expected Uber_XL ROC Curve outperforms that of Uber_X since Uber_XL area under the curve = 1.00, which represents the perfect classifier. What determines a good classifier is what the threshold is set at to classify as 0 or 1. Due to Uber_XL having even data to start with in order to train its logistic regression model, its ability to correctly classify unseen data would be better than Uber_X's dataset that has a huge skewing towards non-cancellations. Therefore, Uber_X's ROC Curve area is 0.5, which is the baseline.

In conclusion, if you want a strong model to predict whether or not fares set are going to lead to cancellations or not, the Uber_XL Logistic Regression model would be the best one to use. This model could be used in real-time to see what pricing strategy response would be. If price adjustments need to be made to increase completed trips then changes can be made more swiftly in order to not lose out on ROI.

If you would like pricing data that leads to less cancellations, Uber_X would be the best data set to use as reference to where to set pricing since the data itself has a high skewing towards non-cancellations.

Therefore, even though Uber_XL logisic regression model outperformed Uber_X's, Uber_X had the better pricing strategy. So Uber_XL will need to reset its pricing to capitalize more on interested Uber riders.
